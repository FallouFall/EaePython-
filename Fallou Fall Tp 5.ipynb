```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

with Image.open("messi.jpg") as messiImage:
    messiPrime = np.array(messiImage)
print(messiPrime.shape)
messiCopy = messiPrime.copy()
x1 = -150
x2 = 180
y1 = 350
y2 = -50
messiCopy[y1:y2, x1:x2, :] = 0
plt.imshow(messiCopy)
plt.show()
```
